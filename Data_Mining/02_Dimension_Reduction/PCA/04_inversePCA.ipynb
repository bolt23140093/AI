```python
from sklearn.decomposition import PCA
import numpy as np

# Sample data
X = np.array([[1, 2, 3], [5, 4, 3], [19, 15, 10]])

# Perform PCA and select the number of components to retain
n_components =1
pca = PCA(n_components=n_components)
X_reduced = pca.fit_transform(X)

# Inverse transform to reconstruct the original data
X_original = pca.inverse_transform(X_reduced)

print("Original data:")
print(X)
print("X_reduced data:")
print(X_reduced)
print("Reconstructed data:")
print(X_original)
```
