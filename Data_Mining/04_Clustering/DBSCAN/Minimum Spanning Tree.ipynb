```python
# -*- coding: utf-8 -*-
"""
Created on Thu Dec 26 09:36:02 2024

@author: joseph@艾鍗學院 www.ittraining.com.tw

Implements Prim's algorithm for finding Minimum Spanning Tree

"""


G = [
    [0, 4, 0, 3, 5],
    [4, 0, 2, 0, 0],
    [0, 2, 0, 1, 0],
    [3, 0, 1, 0, 0],
    [5, 0, 0, 0, 0]
    ]

'''
G = [
    [0, 4,  5,  0,  0, 0],    # A's edges: B(4), C(5)
    [4, 0, 11,  9,  7, 0],    # B's edges: A(4), C(11), D(9), E(7)
    [5,11,  0,  0,  3, 0],    # C's edges: A(5), B(11), E(3)
    [0, 9,  0,  0, 13, 2],    # D's edges: B(9), E(13), F(2)
    [0, 7,  3,  13, 0, 6],    # E's edges: B(7), C(3), D(13), F(6)
    [0, 0,  0,  2,  6, 0]     # F's edges: D(2), E(6)
]
'''


def prims_mst(graph):

    num_vertices = len(graph)

    # Initialize sets as per algorithm
    V_new = {0}  # Start a set with vertex 0
    E_new = []   # Empty set of edges

    # Continue until V_new contains all vertices
    while len(V_new) != num_vertices:
        min_cost = float('inf')
        min_edge = None

        # Find the lowest cost edge (u,v) where u ∈ V_new and v ∈ V - V_new
        for u in V_new:
            for v in range(num_vertices):
                if v not in V_new and graph[u][v] != 0:  # Check if edge exists
                    if graph[u][v] < min_cost:
                        min_cost = graph[u][v]
                        min_edge = (u, v)


        # Add the edge and vertex to respective sets
        u, v = min_edge
        E_new.append(min_edge)
        V_new.add(v)

    # Calculate total cost
    total_cost = sum(graph[u][v] for u, v in E_new)

    return E_new, total_cost

mst_edges, total_cost = prims_mst(G)
print("Edges in the Minimum Spanning Tree:")
for edge in mst_edges:
    print(f"Edge: ({edge[0]}, {edge[1]}), Weight: {G[edge[0]][edge[1]]}")

print(f"\nTotal MST Cost: {total_cost}")
```
