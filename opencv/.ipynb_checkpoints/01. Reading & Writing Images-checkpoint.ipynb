# 5. Contrast and Brightness

```python
img_path = os.path.join('data', 'mask.jpg')
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def contrast_brightness(img, a=1.0, b=0.0):
    alpha = np.array([a]) # contrast 0~3
    beta = np.array([b]) # brightness + or -
    img = cv2.multiply(img, alpha) #增加對比
    img = cv2.add(img, beta) #增加亮度
    img = np.clip(img, 0, 255) #超果255就保持255
    return img
```

```python
# Show image
plt.imshow(contrast_brightness(img))
plt.show()
```

```python
# Show image
plt.imshow(contrast_brightness(img, 2.0, 0.0))
plt.show()
```

```python
# Show image
plt.imshow(contrast_brightness(img, 1.0, 50.0))
plt.show()
```
